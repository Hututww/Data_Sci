## 分形几何

> *无边的奇迹源自简单规则的无限重复。*
>
>  &emsp;&emsp;&emsp;&emsp;*——分形之父 Benoît B. Mandelbrot*

分形（Fractal）一词是数学家 Benoît B. Mandelbrot(1924－2010) 创造出来的，此词源于拉丁文fractus，意为不规则、支离破碎等。

即使这个名字对你来说很陌生，你之前可能也看过的曼德布洛特集合的可视化。

它是一组复数，当在复平面上描绘时，其边界形成独特而复杂的图案。

这种模式可以说是最著名的分形，在20世纪后期催生了分形几何：

![](https://wiki.swarma.org/images/6/6b/P19Mandelbrot_zoom.gif)

分形在自然界也是一种相当普遍的现象。

例如，这种罗马花椰菜具有自相似的结构，它的每个部分看起来都像整体，只是更小：

![](https://article.biliimg.com/bfs/article/f1684daa1fa46941aec180c7257147819f079024.png@1e_1c.webp)

曼德布洛特集合最初是由一位名叫 Mandelbrot 的数学家发现的。

## 分形的重要特征

### 自相似性(Self-similarity)

局部与整体在形态、功能、信息、时间、空间等方面具有统计意义上的相似性，比如树干和枝桠的自相似，罗马花椰菜的整体和局部的自相似等等。

### 无标度性（Scale-free）

当我们变换不同的尺度，都会发现系统还是那样，它没有变化（即自相似性），也就是标度不变性。比如大漩涡套着小旋涡，它们发生在不同的尺度，但是却表现出相似性，这就称为无标度特性。

### 非线性（Nonlinear）

非线性和线性相对，线性是指成比例（可由线性函数 `y = kx + b` 刻画），而非线性则指输入和输出不成比例（可由非线性函数刻画），比如抛物线就是非线性的。在复杂系统中，非线性是最重要的特性之一。


## 曼德布洛特集合

形式上，曼德布洛特集是一个无限的复数集合，这些复数满足在给定迭代条件下，使序列 `z0,z1, ...,zn, ...` 不发散的性质。

曼德布洛特序列由以下递归公式给出（你可以看到它是非线性的）：

![](https://article.biliimg.com/bfs/article/4b5446efa19dc02cf641032bf9c3e46f4f69fffa.png@1e_1c.webp)

也就是说，要确定某个复数`c`是否属于曼德布洛特集合，只需将该数字代入到上面的公式中，然后不断地计算`zn`。

序列的第一个元素，`z0`，始终等于零。计算下一个元素，`z1 = c`，继续平方`z1`并加上`c`，得到`z2 = c² + c`，以此类推。

但是我们不会（也无法）无限迭代下去，当`n`足够大时，我们就能观察到序列是否收敛了，也即序列是否趋于一个稳定值。

如果序列没有发散，那么我们就可以说`c`属于曼德布洛特集合；反之，`c`就不属于曼德布洛特集合。

要绘制图像，我们还需要一张“白纸”，我准备在自定义的绘图函数中实现它：

````python
import numpy as np
import pylab as pl
from matplotlib import cm


def draw_set(cx, cy, d, ufunc: np.ufunc):
    x0, x1, y0, y1 = cx - d, cx + d, cy - d, cy + d
    y, x = np.ogrid[y0:y1:400j, x0:x1:400j] # 这一行将整张白纸分为了`400*400`的小格(像素点)，`400j`表示把`y0`到`y1`(或`x0`到`x1`)平均分为400份。
    z = x + y * 1j
    pl.imshow(ufunc(z).astype(float), cmap=cm.jet, extent=[x0, x1, y0, y1])
````

#### 准备“画纸”

`(cx, cy)`是这张“白纸”的中心，`d`是从中心向两边延伸的长度，也就是说，这张“白纸”是一个以`(cx, cy)`为中心，边长为`2d`的正方形区域。

`y, x = np.ogrid[y0:y1:400j, x0:x1:400j]` 这一行将整张白纸分为了`400*400`的小格(像素点)，`400j`表示把`y0`到`y1`(或`x0`到`x1`)平均分为400份。

可以从下面的示例代码中看出`np.ogrid`是如何起作用的:

````python
y,x = np.ogrid[-1:1:5j, -1:1:5j]
print(x)
print(y)
````

接下来构造了复数`z = x + y*1j`，将这张白纸与我们的复平面联系起来，只用一个复数数组，就能表示这些`x*y`。

在这里，因为`x`的形状是`1*5`，`y`的形状是`5*1`，因此`z`的形状是`5*5`：

````python
z = x + y * 1j
print(z) # 5*5的矩阵
````

我们在实际绘图的时候，为了让图片更清晰，`z`的大小是`400*400`。

#### 让迭代函数直接支持数组运算

`pl.imshow(ufunc(z).astype(float), cmap=cm.jet, extent=[x0, x1, y0, y1])` 这一行是绘图的主要代码。

`ufunc`是一个`np.ufunc`类型的函数，能够方便地处理数组类型的参数。`cmap`是配色方案，我选择的是`jet`。`extent`是绘图区域。

具体来说，我们前面定义的`iter_m(c)`函数，只能处理单个数字，无法直接处理数组，而现在的`z`就是一个复数数组。

为了将一个普通的`Python`函数变为一个可以直接处理数组的函数，我们可以使用`numpy`提供的`frompyfunc(func, nin, nout) -> ufunc`方法。

其中func是我们自定义的函数，`nin`是传入参数的个数，`nout`是传出结果的个数。

具体使用方法如下：

````python
mandelbrot = np.frompyfunc(iter_m, 1, 1)
mandelbrot(z)
````

可以看到，我们成功地将`z`的每一个坐标都判断出了是否属于曼德布洛特集合。

我们画出这个只有5*5像素的图片看看：

````python
pl.imshow(mandelbrot(z).astype(float), cmap=cm.jet, extent=[-1, 1, -1, 1])
````

嗯，果然非常的抽象，看来`400*400`的分辨率是很有必要的。

现在，万事具备，只欠输出。准备好画出我们的第一张曼德布洛特集合的图像了。

````python
draw_set(-0.5, 0, 1.0, mandelbrot)
````

## 茱莉亚集合

讲完了曼德布洛特和他的集合，不得不谈到另一个人，法国数学家Gaston Julia。

他在没有计算机的帮助下发现了Julia集合。

Julia集合和曼德布洛特集合密切相关，因为它们的**递归公式是相同的，只是起始条件不同**。

回顾一下曼德布洛特集合的递归方式：

![](https://article.biliimg.com/bfs/article/4b5446efa19dc02cf641032bf9c3e46f4f69fffa.png@1e_1c.webp)

我们发现`z0`总是`0`，而`c`是可变的参数。

而要确定一个复数`z`是否属于Julia集合，必须使用该数字作为序列的起点，并为`c`参数选择另一个确定的值。

曼德布洛特集合只有一个，而Julia集合有无限多个。

我们只需要对曼德布洛特集合的迭代函数进行简单的修改，就可以应用在Julia集合上：

````python
def iter_j(z):
    c = -0.4 + 0.6j
    for i in range(0, 200):  # 迭代200次
        if(abs(z) > 2):
            break # 半径大于2则认为发散
        z = z**2 + c
    return i
````

接下来我们用和前面相同的一套绘图方法，就能得到Julia集合的彩照了：

````python
julia = np.frompyfunc(iter_j, 1, 1)
draw_set(0, 0, 1.5, julia)
````

同样可以通过一段逐级放大的代码来看看Julia集的细节：

````python
x, y = 0.5754, 0.2048
pl.subplot(231)
draw_set(0, 0, 1.5, julia)
for i in range(2, 7):
    pl.subplot(230 + i)
    draw_set(x, y, 0.25 ** (i - 1), julia)
pl.subplots_adjust(0.05, 0.1, 0.95, 1, 0.5, 0.1)
````
